```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.random.uniform(-3, 3, size=100)
X = x.reshape(-1, 1)
y = 0.5 * x**2 + x + 2 + np.random.normal(0, 1, size=100)
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(degree=2)  # 指定添加2次幂
poly.fit(X)
X2 = poly.transform(X)
```

```python
X2.shape
```

```python
X2[:5,:]  # 第一列为X的0次方， 第二列是X的1次方， 第三列是X的2次方
```

```python
X[:5,:]
```

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X2, y)
X2_predict = lin_reg.predict(X2)
```

```python
plt.scatter(x, y)
plt.plot(np.sort(x), X2_predict[np.argsort(x)], color='r')
```

```python
lin_reg.coef_
```

```python
lin_reg.intercept_
```

## 关于PolynomialFeature

```python
X = np.arange(1, 11).reshape(-1, 2)
```

```python
X.shape
```

```python
X
```

```python
poly = PolynomialFeatures(degree=2)
poly.fit(X)
X2 = poly.transform(X)
```

```python
X2 
# 因为X有两个维度，polynomial features实际上是将X从0到2次方所有可能的组合都hstack起来了
# 第一列为X的0次方， 第二列为X的第一个维度的1次方，第三列是X第二个维度的1次方，
# 第四列是X的第一个维度的2次方，第六列是X第二个维度的2次方
# 第五列是X的两个维度相乘的结果
```

## Pipeline

```python
x = np.random.uniform(-3, 3, size=100)
X = x.reshape(-1, 1)
y = 0.5 * x**2 + x + 2 + np.random.normal(0, 1, size=100)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

# 管道的好处就在于可以直接指定每一步的类，由管道自动化执行，而不用手动敲打每一步的代码，通过管道，我们创建了一个多项式回归器
poly_reg = Pipeline([
                    ('poly', PolynomialFeatures(degree=2)),
                    ('std_scaler', StandardScaler()),
                    ('lin_reg', LinearRegression())
                    ], verbose=True)
```

```python
poly_reg.fit(X, y)
```

```python
y_predict = poly_reg.predict(X)
```

```python
plt.scatter(x, y)
plt.plot(np.sort(x), y_predict[np.argsort(x)], color='r')
```
