><u>**模型误差 = 偏差(Bias) + 方差(Variance) + 不可避免的误差**</u>

`导致偏差的主要原因：`
- 对问题本身的假设不正确

    如：非线性数据使用线性回归，即欠拟合(underfitting)
    
    
- 使用的特征跟预测的问题没有什么关系

    如：使用学生的名字来预测学生的成绩

`方差：`

高方差体现在，数据一点点的扰动都会较大的影响模型。

通常原因，使用的模型太复杂，如高阶多项式回归，出现的过拟合现象(overfitting)。

***

***注意：***

有一些算法天生就是高方差的算法，如KNN。

非参数学习通常是高方差算法，因为不对数据进行任何的假设。

有一些算法天生就是高偏差算法，如线性回归。

参数学习通常是高偏差算法，因为对数据具有极强的假设。

大多数算法具有相应的参数，可以调整偏差和方差，如KNN中的K，如线性回归中使用多项式回归。


<u>**偏差和方差通常是矛盾的。**

**降低偏差，会提高方差。**

**降低方差，会提高偏差。**</u>

## 方差

**其实，机器学习的主要挑战，来自方差。即过拟合(Overfitting)问题**

降低高方差的通常手段：
1. 降低模型复杂度
2. 降低数据维度；降噪
3. 增加样本数
4. 使用验证集

5. <u>模型正则化</u>
