如果只使用一份训练数据集和测试数据集，有可能会出现在特定测试数据集上的过拟合现象。因为我们模型一直在围绕这一份测试数据集打转，最终当我们选择出对测试数据集拟合效果最好的模型时，有可能已经在测试数据集上出现过拟合情况而不自知。

为了解决这个问题，我们需要将数据集分为训练数据集、验证数据集和测试数据集。训练数据集依然用来训练，验证数据集则是承担了之前测试数据集的工作，用来验证模型的效果，如果效果不好，再进行调参。最后，对最终模型使用测试数据集进行最终的效果测试。

因此测试数据集实际上是作为衡量最终模型性能的数据集。而验证数据集是作为调整超参数使用的数据集。

## 问题：随机

由于验证数据集也是只有一份，并且是随机筛选的，如果验证数据集中有一些极端值，有可能导致模型对验证数据集过拟合。因此我们需要引进交叉验证(Cross validation)

## 交叉验证 Cross Validation

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
```

```python
digits = datasets.load_digits()
X = digits.data
y = digits.target
```

### 测试train_test_split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=666)
```

```python
from sklearn.neighbors import KNeighborsClassifier

# 调参
best_score, best_k, best_p= 0, 0, 0
for k in range(2,11):
    for p in range(1,6):
        knn_clf = KNeighborsClassifier(weights='distance', n_neighbors=k, p=p)
        knn_clf.fit(X_train, y_train)
        score = knn_clf.score(X_test, y_test)
        if score > best_score:
            best_score, best_k, best_p = score, k, p
            
print("Best k = :", k)
print("Best p = :", p)
print("Best score = :", score)
```

### 使用交叉验证

```python
from sklearn.model_selection import cross_val_score

knn_clf = KNeighborsClassifier()
cross_val_score(knn_clf, X_train, y_train)
```

```python
from sklearn.neighbors import KNeighborsClassifier

best_score, best_k, best_p = 0, 0, 0

for k in range(2, 11):
    for p in range(1, 6):
        knn_clf = KNeighborsClassifier(weights='distance', n_neighbors=k, p=p, n_jobs=-1)
        scores = cross_val_score(knn_clf, X_train, y_train, n_jobs=-1)
        score = np.mean(scores)
        if score > best_score:
            best_score, best_k, best_p = score, k, p

print("Best k = :", best_k)
print("Best p = :", best_p)
print("Best score = :", best_score)
```

```python
best_knn_clf = KNeighborsClassifier(weights='distance', n_neighbors=2, p=2)
```

```python
best_knn_clf.fit(X_train, y_train)
best_knn_clf.score(X_test, y_test)
```

### 总结

把训练数据集分成k份，称为k-fold cross validation

交叉验证的缺点：每次训练k个模型，相当于整体性能慢了k倍。

### 留一法 LOO_CV

把训练数据集分成m份，每次将m-1份样本用于训练，将剩下的1份作为验证数据集，称为留一法

Leave-One-Out Cross Validation


这样做将完全不受随机的影响，最接近模型真正的性能指标

缺点：计算量巨大
