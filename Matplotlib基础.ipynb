```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

## 绘制折线图

```python
x= np.linspace(0,10,100)
siny= np.sin(x)
cosy= np.cos(x)
tany= np.tan(x)
```

```python
plt.plot(x,siny,c= 'b',linestyle= "--",label= "sin(x)")
plt.plot(x,cosy,c= 'r',linestyle= "-.",label= "cos(x)")
plt.xlim(-1,12)
plt.ylim(-1.5,1.5)
plt.xlabel("x axis")
plt.ylabel("y value")
plt.legend(loc= 3) ##图例
plt.title("First step") ##标题
plt.show()
```

```python
plt.plot(x,tany,c= 'black')
plt.axis([-1,11,-45,25])
```

## 绘制散点图（Scatter plot)

```python
plt.scatter(x,siny,s= 0.5,c= "red")
```

```python
x= np.random.normal(loc=0,scale=1,size= 10000) #创建一组服从正态分布的随机数，其中均值为0，方差为1，即标准正态分布
y= np.random.normal(0,1,10000) #分布与x一致
plt.scatter(x,y,alpha=0.1) #alpha为透明度
plt.show()
```
