```python
from sklearn.decomposition import PCA
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python

X = np.empty((100,2))
X[:,0] = np.random.uniform(0., 100., size=100)
X[:,1] = 0.75 * X[:, 0] + 3. + np.random.normal(0., 10., size=100)
```

```python
pca = PCA(n_components=1)
pca.fit(X)
```

```python
pca.components_
```

```python
X_reduction = pca.transform(X)
X_restore = pca.inverse_transform(X_reduction)
```

```python
plt.scatter(X[:,0], X[:,1], color='r', alpha=0.5) # X
plt.scatter(X_restore[:,0], X_restore[:,1], color='b') # 恢复后的X
plt.show()
```

## 使用真实数据集

```python
from sklearn import datasets
digits = datasets.load_digits()
```

```python
print(digits.DESCR)
```

```python
X = digits.data
y = digits.target
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 666)
```

```python
X_train.shape
```

```python
%%time
from sklearn.neighbors import KNeighborsClassifier
knn_clf = KNeighborsClassifier()
knn_clf.fit(X_train, y_train)
```

```python
knn_clf.score(X_test, y_test)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
pca.fit(X_train)
X_train_reduction = pca.transform(X_train)
X_test_reduction = pca.transform(X_test)
```

```python
%%time
knn_clf = KNeighborsClassifier()
knn_clf.fit(X_train_reduction, y_train)
```

```python
knn_clf.score(X_test_reduction, y_test)
```

**在将原有的64维数据直接降维至2维后，数据损失导致分类准确度直接从原来的98.67% 下降到 60.67% ，这个结果我们不能接受，尝试将降维的维度进行调整**

```python
# pca解释方差比例，越大说明能解释的原数据的程度越高，目前两个维度相加仅能解释原数据的28%左右，其余的78%的信息已经丢失了
pca.explained_variance_ratio_
```

### 寻找最佳的降维维度

```python
pca = PCA(n_components=X_train.shape[1])
pca.fit(X_train, y_train)
pca.explained_variance_ratio_
```

```python
plt.plot([i for i in range(X_train.shape[1])], \
          [np.sum(pca.explained_variance_ratio_[:i + 1]) for i in range(X_train.shape[1])])
plt.show()
```

可以看出超过30维后，降维后的数据解释方差比例已经趋近于1，因此我们可以采取大于30的数作为n_components的值。

其实这种通过解释方差比例来确定降维维数的方式，sklearn已经替我们封装好了，用法如下

```python
pca = PCA(0.95)  # 让算法自己去选择解释方差比例在95%的维数
pca.fit(X_train)
```

```python
pca.n_components_  
# 28维
```

```python
X_train_reduction = pca.transform(X_train)
X_test_reduction = pca.transform(X_test)
```

```python
knn_clf = KNeighborsClassifier()
knn_clf.fit(X_train_reduction, y_train)
knn_clf.score(X_test_reduction, y_test)
```

**那是不是降到2维就毫无意义呢？并不是，降到2维有一个最直接的好处，就是可视化**

```python
pca = PCA(n_components=2)
pca.fit(X, y)
X_reduction = pca.transform(X)
```

```python
digits.target_names
```

```python
plt.figure(figsize=(12,6))
for i in range(10):
    plt.scatter(X_reduction[y == i,0], X_reduction[y == i, 1], alpha= 0.8)
plt.show()
```
